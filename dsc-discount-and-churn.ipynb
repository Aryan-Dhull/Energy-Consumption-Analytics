```python
import pandas as pd
import scipy.stats as stats
import numpy as np
from scipy.stats import poisson
import math
```

```python
df = pd.read_csv("/kaggle/input/project-dataset/data_for_predictions (1).csv")
df
```

```python
def mean(arr):
    return sum(arr)/len(arr)
```

```python
def variance(arr):
    return sum((arr-mean(arr))**2)/len(arr)
```

```python
def correlation_coefficient(X,Y):
    mean_X = mean(X)
    mean_Y = mean(Y)
    
    num = sum((X-mean(X))*(Y-mean(Y)))
    den = (sum((X-mean(X))**2))*(sum((Y-mean(Y))**2))
    
    return num/(den)**0.5
```

```python
r=correlation_coefficient(df['churn'],df['forecast_discount_energy'])
n=14606
r
```

```python
def t_statistic(r,n):
    return (r/(1-(r)**2)**0.5)*((n-2)**0.5)
```

```python
t = t_statistic(r,n)
alpha = 0.01
critical_value = stats.t.ppf(1 - alpha / 2, n-2)
```

```python
print(f't-statistic: {t}')
print(f'Critical value will be: {critical_value}')

if abs(t) > critical_value:
    print('Reject the null hypothesis as there is a significant correlation.')
else:
    print('Fail to reject the null hypothesis.')
```
