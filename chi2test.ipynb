```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load data
client_data = pd.read_csv('/kaggle/input/dsc-proj-dataset/data_for_predictions.csv')

# Create a binary variable: single vs. multiple products
client_data['multiple_products'] = (client_data['nb_prod_act'] > 1).astype(int)

# Create contingency table
product_churn_table = pd.crosstab(client_data['multiple_products'], client_data['churn'])
# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(product_churn_table)

print("Product Variety and Churn")
print(f"Chi2 Statistic: {chi2}, p-value: {p}")
```
